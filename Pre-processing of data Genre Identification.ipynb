```python
nan = []
for i in range(len(data)):
    c=0;
    for j in range(1,12):
        a='Genre'+str(j)
        loc=data.columns.get_loc(a) 
        if (pd.isna(data.iloc[i,loc])):
            c=c+1
            
    if c==11:
        nan.append(i)
for i in range(len(nan)):
    data=data.drop(nan[i],axis=0)
```

```python
data = data[data['ID'].notna()] #rows that have nan values in id is delted.
```

```python
data = data[data['Author/Publisher'].notna()] #rows that have nan values in id is delted.
```

```python
data = data[data['Title'].notna()] #rows that have nan values in id is delted.
```

```python
data = data[data['Summary'].notna()] #rows that have nan values in id is delted.
```

```python
data = data.dropna(axis = 0, how = 'all') #rows that have all NaN values
```

```python
for i in range(1,12):
    a='Genre'+str(i)
    data[a] = data[a].fillna(value="") #replacing the Genre NaN values with empty strings
```
